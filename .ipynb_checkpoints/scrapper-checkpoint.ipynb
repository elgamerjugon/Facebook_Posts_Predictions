```python
url_reforma = ["https://www.reforma.com/rss/portada.xml",
               "https://www.reforma.com/rss/ciudad.xml",
               "https://www.reforma.com/rss/justicia.xml",
               "https://www.reforma.com/rss/estados.xml",
               "https://www.reforma.com/rss/negocios.xml",
               "https://www.reforma.com/rss/internacional.xml",
               "https://www.reforma.com/rss/cancha.xml",
               "https://www.reforma.com/rss/cultura.xml",
               "https://www.reforma.com/rss/ciencia.xml",
               "https://www.reforma.com/rss/nacional.xml",
               "https://www.reforma.com/rss/editoriales.xml",
               "https://www.reforma.com/rss/gadgets.xml",
               "https://www.reforma.com/rss/moda.xml",
               "https://www.reforma.com/rss/estados.xml",
               "https://www.reforma.com/rss/de%20viaje.xml"
              ]

url_excelsior = ["https://www.excelsior.com.mx/rss.xml"]

url_eluniversal = ["http://www.eluniversal.com.mx/rss/universalmxm.xml",
                   "http://www.eluniversal.com.mx/rss/mundo.xml",
                   "http://www.eluniversal.com.mx/rss/mexico.xml",
                   "http://www.eluniversal.com.mx/rss/estados.xml",
                   "http://www.eluniversal.com.mx/rss/finanzas.xml",
                   "http://www.eluniversal.com.mx/rss/espectaculos.xml",
                   "http://www.eluniversal.com.mx/rss/cultura.xml",
                   "http://www.eluniversal.com.mx/rss/computo.xml",
                   "http://www.eluniversal.com.mx/rss/sociedad.xml",
                  ]

def read_entries(url):
    df = pd.DataFrame()
    if len(url) > 1:
        for link in url:
            print(f"Leyendo las noticias del siguiente RSS: {link}")
            reader = feedparser.parse(link)
            df = df.append(reader.entries)
    else:
        print(f"Leyendo las noticias del siguiente RSS: {url[0]}")
        reader = feedparser.parse(url[0])
        df = df.append(reader.entries)
        
    print(f"Se ha terminado de leer el RSS satisfactoriamente\n")
    return df

reforma_df = read_entries(url_reforma)
excelsior_df = read_entries(url_excelsior)
eluniversal_df = read_entries(url_eluniversal)
```

```python
def data_cleaning(df):
    df["id"] = list(range(len(df)))
    df.reset_index(inplace = True)
    df.set_index(df.id, inplace = True)
    df["summary"] = df["summary"].str.replace(r"\<.+\> ", "")
    df.replace("", np.nan, inplace = True)
    df.dropna(inplace = True)
    return df

eluniversal_df = data_cleaning(eluniversal_df)
reforma_df = data_cleaning(reforma_df)
excelsior_df = data_cleaning(excelsior_df)
```

```python
def remove_columns(df):
    cols_names_remove = ["index", "idelementosubcategoria", "title_detail", "links", "id", "guidislink",
                         "grlink", "published_parsed", "summary_detail", "authors", "author_detail", "textofacebook",
                         "textotwitter", "idcategoria", "idsubcategoria"] 
    cols_remove = []
    
    for cols in df.columns:
        if cols in cols_names_remove:
            cols_remove.append(cols)
            
    df.drop(columns = cols_remove, inplace = True)
    return df
```

```python
reforma_df = remove_columns(reforma_df)
eluniversal_df = remove_columns(eluniversal_df)
excelsior_df = remove_columns(excelsior_df)
```
