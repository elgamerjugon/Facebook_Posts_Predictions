```python
translations = pd.read_csv("translations.csv")
translations.drop(columns = ["Unnamed: 0"], inplace = True)
translations.rename(columns = {"0" : "translated"}, inplace = True)
annual_df["translated"] = translations.translated
annual_df
```

```python
analyzer = SentimentIntensityAnalyzer()

def sentiment_analyzer_scores(sentence):
    score = analyzer.polarity_scores(sentence)
    return score
```

```python
# annual_df[annual_df.score == ""]

annual_df_sliced = annual_df[0:533]
annual_df_sliced.reset_index(inplace = True)
annual_df_sliced
```

```python
annual_df_sliced.drop(columns = ["index"], inplace = True)

annual_df_sliced["score"] = ""
for i, new in enumerate(annual_df_sliced["translated"]):
    dict_score = sentiment_analyzer_scores(new)
    annual_df_sliced["score"][i] = dict_score["compound"]
```

```python
def sentiment(score):
    if score >= 0.05:
        # Positive sentiment
        return 3
    elif score > -0.05 and score < 0.05:
        # Neutral 
        return 2
    else:
        #Negative
        return 1
```

```python
annual_df["sentiment"] = annual_df.score.apply(lambda x: sentiment(x))
annual_df
```

```python
ml_df = annual_df_sliced.drop(columns = ["created", "url", "message", "link", "description", "translated", "page_name", "user_name"])
```

```python
ml_df.overperforming_score = ml_df.overperforming_score.apply(lambda x: x.replace(",", ""))
```

```python
ml_df.overperforming_score = pd.to_numeric(ml_df.overperforming_score, downcast = "float")
ml_df.score = pd.to_numeric(ml_df.score, downcast = "float")
ml_df.info()
```

# Applying ML Model

```python
X = ml_df.drop(columns = ["overperforming_score"])
y = ml_df.overperforming_score
```

```python
linereg = LinearRegression(fit_intercept = True)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
linereg.fit(X_train, y_train)
```

```python
y_hat = linereg.predict(X_test)
y_hat
```

```python
rmse = (mean_squared_error(y_test, y_hat))**(0.5)
rmse
linereg.score(X, y)
```

```python
# Crea un problema de regresión aleatorio
# X, y = make_regression(n_features=9, n_informative=2,
#                        random_state=0, shuffle=False)
regr = RandomForestRegressor(max_depth=10, random_state=0,
                             n_estimators=100)

scores = cross_validate(regr, X, y, cv=5,
                        scoring=('r2', 'neg_mean_squared_error'),
                        return_train_score=True)
regr.fit(X_train, y_train)
y_hat_random = regr.predict(X_test)
rmse = (mean_squared_error(y_test, y_hat_random))**(0.5)
# np.mean((scores["test_neg_mean_squared_error"]*-1)**0.5)
rmse
```

```python
regr.feature_importances_
regr.score(X, y)
```

# Applying Bagging Regressor

```python
bagging_regressor = BaggingRegressor()
bagging_regressor.fit(X_train, y_train)
y_hat_bagging = bagging_regressor.predict(X_test)
rmse = (mean_squared_error(y_test, y_hat_bagging))**(0.5)
# rmse
scores = cross_validate(bagging_regressor, X, y, cv=5,
                        scoring=('r2', 'neg_mean_squared_error'),
                        return_train_score=True)
np.mean((scores["test_neg_mean_squared_error"]*-1)**0.5)
bagging_regressor.score(X, y)
```
